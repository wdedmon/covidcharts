```
import pandas as pd
import numpy as np
import requests
import csv
import seaborn as sns
import matplotlib as plt
```

```
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv')
```

```
df.head()
```

```
import seaborn as sns

palette = sns.light_palette("seagreen")

g1 = sns.relplot(
    data=df,
    x="date", y="total_vaccinations_per_hundred",
    hue="location", col="location",
    kind="line",
    height=3, aspect=1, facet_kws=dict(sharex=False), col_wrap=5, 
)
```
